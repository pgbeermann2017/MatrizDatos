```
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
import numpy as np

grid_x, grid_y = np.mgrid[-1:1:100j, -1:1:200j]

x1=np.linspace(-1,1,40)
y1=np.linspace(-1,1,40)

v=[]
points=[]
for x2 in x1:
  for y2 in y1:
      pts=np.array([x2,y2,0])
      
      v.append(my_interpolating_function(pts)[0])
      points.append([x2,y2])
      
values=np.array(v)  
points=np.array(points)
print(values.shape,points.shape)

grid_z2 = griddata(points, values, (grid_x, grid_y), method='cubic')

plt.figure(figsize=(12,12))
#plt.imshow(grid_z2, extent=(-1,1,-1,1), origin='lower')

contours=plt.contour(grid_z2, 20, cmap='RdGy');
plt.clabel(contours, inline=True, fontsize=8)
plt.colorbar();
plt.axis(aspect='image')
plt.show()
```
